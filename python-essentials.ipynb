# <b>Flow control</b>

## if - else

```python
num = float(input("Enter a number: "))
if num > 0:
    print("Your number is positive")
elif num < 0:
    print("Your number is negative")
else:
    print("Your number is zero")
```

## for loop

```python
original_strings = ['house', 'computer', 'book', 'notebook']   
```

```python
strings_with_their_size = []
for s in original_strings:
    new_s = s + '_' + str(len(s))
    strings_with_their_size.append(new_s)
```

```python
strings_with_their_size
```

```python
[s + '_' + str(len(s)) for s in original_strings]
```

```python
for n in range(3):
    print(n)
```

```python
for i, n in enumerate([11, 13, 17, 19]):
    print(f"Index: {i}, number: {n}")
```

## while loop

```python
num = int(input("Enter a number: "))
iteration = 1
while num < 1_000_000_000 and iteration <= 20:
    print(f"Iteration {iteration} - num is {num}")
    num = num ** 2
    iteration += 1
```

## try - catch

```python
n = int(input("Enter this number -> 4242: "))
if n != 4242:
    raise ValueError("You entered the wrong number")
```

```python
123 / 0
```

```python
n1 = int(input("Enter numerator: "))
n2 = int(input("Enter denominator: "))
try:
    division_result = n1 / n2
except ZeroDivisionError:
    print("You should not divide by 0")
else:
    print(f"The result is {division_result}")
finally:
    print(f"End of program")
```

# <b>Functions</b>

```python
def capitalize_first_and_last_letter(word):
    first_letter = word[0]
    last_letter = word[-1]
    letters_in_the_middle = word[1:-1]
    new_word = first_letter.upper() + letters_in_the_middle + last_letter.upper()
    return new_word
```

```python
w = 'street'
cap_w = capitalize_first_and_last_letter(word=w)
print(cap_w)
```

# <b>Interactive graphs</b>

```python
# plotly is not in python standard library, must be installed with 'pip install plotly'
import plotly.express as px
```

```python
# generate random samples from a uniform distribution between -1, 1
samples_uniform_distribution = [random.uniform(-1, 1) for x in range(10_000)]
```

```python
# scatter plot (points)
px.scatter(samples_uniform_distribution)
```

```python
# histogram plot (distribution)
px.histogram(samples_uniform_distribution)
```

```python
# generate random samples from a gaussian distribution with mean 10 and standard deviation 5
samples_gaussian_distribution = [random.gauss(10, 5) for x in range(10_000)]
```

```python
# scatter plot (points)
px.scatter(samples_gaussian_distribution)
```

```python
# histogram plot (distribution)
px.histogram(samples_gaussian_distribution)
```

```python
# simulate a list of temperatures by hour of day
hours_of_day = [datetime.datetime(2021, 7, 1, x, 0, 0) for x in range(23)]
temperatures = [25 + random.uniform(-3, 3) for x in range(23)]
```

```python
# lines plot
px.line(x=hours_of_day, y=temperatures)
```

```python
# bar plot
px.bar(x=victories.keys(), y=victories.values())
```
